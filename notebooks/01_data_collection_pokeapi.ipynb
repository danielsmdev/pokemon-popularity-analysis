### ğŸ“Œ  Extraer Datos Detallados de Cada PokÃ©mon

### Importamos todos los mÃ³dulos

```python
import pandas as pd
import time

from pokemon_api import get_pokemon_names, get_all_forms, get_pokemon_data, get_species_data, clean_pokemon_name, get_base_name
from data_processing import save_partial_data, filter_pokemon_variants
from config import POKEAPI_PARTIAL_SAVE_PATH_DATA_COLLECTION, POKEAPI_SAVE_PATH_DATA_COLLECTION
```

```python
# ğŸ“Œ Obtener lista de PokÃ©mon
pokemon_list = get_pokemon_names()

# ğŸ“Œ Obtener datos con variantes
pokemon_data = []
species_data = []

for i, p in enumerate(pokemon_list, 1):
    print(f"\nğŸ” Obteniendo datos de: {p} ({i}/{len(pokemon_list)})")

    # Obtener todas las formas del PokÃ©mon probando primero con el nombre completo
    all_forms = get_all_forms(p)
    
    for form in all_forms:
        poke_data = get_pokemon_data(form, i)  # Usa el ID base en todas sus variantes
        species_info = get_species_data(p)  # Usamos el nombre original para la especie

        if poke_data and species_info:
            pokemon_data.append(poke_data)
            species_data.append(species_info)

    # Guardado parcial cada 50 PokÃ©mon
    if i % 50 == 0:
        df_partial = pd.DataFrame(pokemon_data)
        df_partial.to_csv(f"{POKEAPI_PARTIAL_SAVE_PATH_DATA_COLLECTION}/pokemon_partial_{i}.csv", index=False)
        print(f"ğŸ“ Guardado parcial en '{POKEAPI_PARTIAL_SAVE_PATH_DATA_COLLECTION}/pokemon_partial_{i}.csv'")

# ğŸ“Œ Convertir a DataFrame
df_pokemon = pd.DataFrame(pokemon_data)
df_species = pd.DataFrame(species_data)

# ğŸ“Œ Fusionar datos
df_pokemon = pd.concat([df_pokemon, df_species], axis=1)

# ğŸ“Œ Guardar el dataset final
df_pokemon.to_csv(POKEAPI_SAVE_PATH_DATA_COLLECTION, index=False)

print("âœ… Dataset COMPLETO guardado correctamente.")
```
